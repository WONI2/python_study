```python
df1 = pd.DataFrame({'CustomerId' : range(1, 11), 'Menu' : ("Kimbab " * 3 + "Bibimbab " * 3 + "Odeng " * 4).split()})
df2 = pd.DataFrame({'CustomerId':[2,4,6],'Table': ("SectorA "*2 + "SectorB").split()})
df3 = pd.DataFrame({'CustomerId':[2,5], 'Gender' : ["Male","Female"]})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
df4 = pd. merge(df1,df2, on = "CustomerId", how = 'outer')
df4
```

```python
df5 = pd.merge(df4, df3, on="CustomerId", how = 'left')
df5
# 아래 나온 결과값에서 메뉴결, 예측치 등을 결정할 수 있음 
```
